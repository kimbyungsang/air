### IPython CheatSheet
* 탭자동완성
* 자기관찰: datetime? b?
* %run 명령어 : %run ipython_script_test.py  <-> 실행중인 코드 중지 Ctrl+c
* 클립보드에 있는 코드 실행하기 : Ctrl+Shift+V, %paste, %cpaste
* 예외와 트레이스백 : %run, %debug, %pdb, %xmode
* 입출력 변수 : _, __, _iX, _X
* 명령어와 별칭 (OS Commands)
* 디렉토리 북마크: %bookmark db /home/wesm/Dropbox/, %bookmark -i
* Interactive Debugger (ipdb) : %debug


### ShortCut
* C(b), C(f), C(a), C(e) : 앞,뒤,맨앞,맨뒤
* C(p, C(n), C(r) : 히스토리 검색
* C(k), C(u), C(l): 삭제하기

### Magic Commands
* %quickref, %magic, %debug, %hist, %pdb (예외발생시 자동으로 디버거 진입), %paste, %cpaste 
* %reset (모든 변수와 이름 삭제), 
* %page OBJECT, 
* %run STATEMENT, %prun (프로파일), %time, timeit
* %who, %who_Is, %who
* %xdel VARIABLE 

### OS Commands
* !cmd, output = !cmd args (명령결과 저장)
* %alias alias_name cmd (명령어 별칭) 
* %cd, %dhist, %pwd, %dirs, %env
